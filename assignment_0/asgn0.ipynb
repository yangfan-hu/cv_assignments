#### 任务2：颜色空间变换（调用OpenCV函数）

**要求：**
1. 将原图像转换为以下颜色空间并显示（使用`cv2.imshow`）：
   - HSV（色相、饱和度、明度）
   - Lab（明度和色彩对立维度）
   - YCrCb（亮度、红色差、蓝色差）
2. 保存转换后的图像为`hsv.jpg`、`lab.jpg`、`ycr.jpg`
3. 从HSV空间中提取并显示H、S、V三个通道（使用3个窗口分别显示） split函数
4. 手动实现图片的灰度化

**原理说明：**
- **HSV颜色空间**：将颜色分解为色调(H)、饱和度(S)和明度(V)，更符合人类描述颜色的方式
- **Lab颜色空间**：感知均匀的颜色空间，L分量表示亮度，a和b表示颜色对立维度
- **YCrCb颜色空间**：常用于图像压缩，Y表示亮度，Cr和Cb表示色差

```python
### YOUR CODE HERE

# 1. HSB
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imwrite('hsv.jpg', hsv_img)
cv2.imshow('HSV Image', hsv_img)

# 2. Lab
lab_img = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2.imwrite('lab.jpg', lab_img)
cv2.imshow('Lab Image', lab_img)

# 3. YCrCb
ycr_img = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
cv2.imwrite('ycr.jpg', ycr_img)
cv2.imshow('YCrCb Image', ycr_img)

# 4. Split HSV
h, s, v = cv2.split(hsv_img)

# show channels
cv2.imshow('Hue Channel (H)', h)
cv2.imshow('Saturation Channel (S)', s)
cv2.imshow('Value Channel (V)', v)

### YOUR CODE END

# pause window
cv2.waitKey(0)
cv2.destroyAllWindows()



# print channel information
print(f"HSV图像形状: {hsv_img.shape}")
print(f"H通道范围: [{np.min(h)}， {np.max(h)}]")
print(f"S通道范围: [{np.min(s)}， {np.max(s)}]")
print(f"V通道范围: [{np.min(v)}， {np.max(v)}]")

```

### 2.4 手动实现灰度化（不使用cvtColor）

**要求：**
1. 根据加权平均法公式实现灰度化：`Gray = 0.299 * R + 0.587 * G + 0.114 * B`
2. 注意OpenCV中颜色通道的顺序为BGR
3. 将手动实现的灰度图像保存为`gray_manual.jpg`
4. 比较手动实现与OpenCV调用（任务1中的`gray_cv.jpg`）的差异
   - 计算两图像的绝对差异
   - 显示差异图像
   - 输出最大差异值和平均差异值

**原理说明：**

灰度化是将彩色图像转换为灰度图像的过程，常用的方法有：

| 方法 | 公式 | 特点 |
|------|------|------|
| 加权平均法 | Gray = 0.299R + 0.587G + 0.114B | 符合人眼灵敏度，最常用 |
| 平均值法 | Gray = (R + G + B) / 3 | 简单但效果较差 |
| 最大值法 | Gray = max(R, G, B) | 亮度偏高 |
| 最小值法 | Gray = min(R, G, B) | 亮度偏低 |

本实验要求使用**加权平均法**手动实现。

```python
def convert_to_gray_scale(bgr_image):
    ### YOUR CODE HERE

    b, g, r = bgr_image[:,:,0], bgr_image[:,:,1], bgr_image[:,:,2]
    gray_image = 0.299 * r + 0.587 * g + 0.114 * b

    ### YOUR CODE END
    return gray_image


img_gray = convert_to_gray_scale(img)

img_gray_ref = np.load('./references/lena_gray.npy')

cv2.imshow('gray_manual', img_gray.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()

if not np.allclose(img_gray, img_gray_ref):
    print('Incorrect values! Please check your implementation.')
```
